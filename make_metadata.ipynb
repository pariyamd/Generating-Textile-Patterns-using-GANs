```python
import pandas as pd
```

```python
import os
```

```python
records=[]
for folder in os.listdir("stylegan2-ada-pytorch/Generated_selection/"):
    for img in os.listdir(f"stylegan2-ada-pytorch/Generated_selection/{folder}"):
        records.append({"name":img,"label":f"{folder}","path":f"{folder}/{img}"})
```

```python
pd.DataFrame.from_records(records).to_csv("Generated_selection.csv")
```
