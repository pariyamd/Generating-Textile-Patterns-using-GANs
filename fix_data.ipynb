```python
import pandas as pd
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
df=pd.read_csv("test_urls.csv", index_col=0)
```

```python
df.drop(["url","name","labels"],axis=1).sum()
```

```python
import glob
import matplotlib.image as mpimg
```

```python
%matplotlib widget

def onpress(event):
    global img_path
    global wrong 
    global change
    global right
    global df
    
    #Get key 
    key = event.key
    
    #Get image name
    path = img_path[0]
    ID = path.split("/")[-1]
    
    
    if key not in ["l","r","w"]:
        print("Invalid Key")
    else: 
        
        if key == "w":
            wrong.append(ID)
            # labels.append("wrong")
            
        elif key == "r":
            right.append(ID)
            # labels.append("right")
            
        elif key == "l":
            change.append(ID)
            # labels.append("change")
        
        # Clear plot 
        plt.clf()

        if len(img_path) > 0: 
            
            img_path = img_path[1:]
            
            #Display next image with count 
            img_name = img_path[0]
            path="fabrics/"+img_name
            img = mpimg.imread(path)
            plt.text(0, 15,list(df[df.name==img_name].labels)[0] ,color='r',size=20)
            
            plt.imshow(img)
            plt.show()
            
global img_path 
global wrong 
global change
global right
global df
wrong = []
change = []
right = []

#Get all image paths
img_path = df.name


fig = plt.figure(figsize=(5,5))

#Load first image and add count
img_name = img_path[0]
path="fabrics/"+img_name
img = mpimg.imread(path)
plt.text(0, 15, list(df[df.name==img_name].labels)[0] ,color='r',size=20)

#Add an interactive widget to figure 
cid = fig.canvas.mpl_connect('key_press_event', onpress)

plt.imshow(img)
plt.show()
```
